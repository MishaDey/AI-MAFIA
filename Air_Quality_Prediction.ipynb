```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
import seaborn
%matplotlib inline
```

```python
def get_data(train_directory_path,test_directory_path):
    training_data=pd.read_csv(train_directory_path)
    #print(training_data.describe())
    x_train = pd.read_csv(train_directory_path).values[0:,:5]
    y_train = pd.read_csv(train_directory_path).values[0:,5:]
    x_test =pd.read_csv(test_directory_path)
    return x_train,y_train,x_test,training_data
```

```python
def train_model(x_train,y_train):
    model=LinearRegression()
    model.fit(x_train,y_train)
    return model
```

```python
def predict_y(model,x_test):
    y_pred=model.predict(x_test)
    return y_pred
```

```python
def variables_correlation(training_data):
    seaborn.heatmap(training_data.corr(),annot=True,cmap='viridis')
    plt.title('Correlation between the variables :\n')
    plt.show()
```

```python
def plot_features_versus_output(data_frame):
    columns=data_frame.columns.tolist()
    y_val=columns[-1]
    for features in columns:
        if(features!=y_val):
            seaborn.lmplot(x=features,y=y_val,data=data_frame)
            plt.show()
```

```python
def main():
    train_directory_path="/home/mishadey/Desktop/AI_MAFIA/3_LINEAR_AND_LOGISTIC_REGRESSION/Assignments (Linear and Logistic Regression)/Linear Regression/Assignment 2/Training Data/Train.csv"
    test_directory_path="/home/mishadey/Desktop/AI_MAFIA/3_LINEAR_AND_LOGISTIC_REGRESSION/Assignments (Linear and Logistic Regression)/Linear Regression/Assignment 2/Testing Data/Test.csv"
    x_train,y_train,x_test,training_data=get_data(train_directory_path,test_directory_path)
    model=train_model(x_train,y_train)
    variables_correlation(training_data)
    print("    Training features v/s Training outputs:")
    plot_features_versus_output(training_data)
    y_pred=predict_y(model,x_test.values)
    y_pred=y_pred.reshape((-1,1))
    x_test['Predicted_Output']=y_pred
    testing_data=x_test
    print("     Testing features v/s Predicted outputs:")
    plot_features_versus_output(testing_data)
    return testing_data
```

```python
if __name__ =='__main__':
    testing_data=main()
```
