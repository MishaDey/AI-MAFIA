```python
import os
import hashlib
import io
from PIL import Image
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from bs4 import BeautifulSoup #web scrapper which will download the entire module and help us to scrap
import requests
import pandas as pd
from pandas import Series,DataFrame
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import presence_of_element_located
```

```python
#function to download the images from a url
#!pip install pillow
#pillow -PIL(python imaging library) fork
def download_image(folder_path,url):
    image_content=requests.get(url).content
    image_file=io.BytesIO(image_content)
    image=Image.open(image_file).convert('RGB')
    file_path=os.path.join(folder_path,hashlib.sha1(image_content).hexdigest()[:10]+'.jpg')
    with open(file_path,'wb') as file:
        image.save(file,"JPEG",quality=85)
    print("SUCCESS -saved {url1} -as {file_path1}".format(url1=url,file_path1=file_path))
```

```python
url='https://www.flipkart.com/'#input("Enter the url:\n")
query='mobile'#input("Enter the search query:\n")
web_driver=webdriver.Chrome(ChromeDriverManager().install())
#result=requests.get(url)
with web_driver as driver:
    wait=WebDriverWait(driver,10) # driver time out
    driver.get(url)
    login_pop = driver.find_element_by_class_name('_29YdH8')
    login_pop.click()
    search_field = driver.find_element_by_class_name('LM6RPg')
    search_field.send_keys(query + "\n")
    wait.until(presence_of_element_located((By.CLASS_NAME,"_31qSD5")))
    content=driver.page_source
    soup=BeautifulSoup(content)
    product_names=[]
    product_imgs=[]
    product_ratings=[]
    product_prices=[]
    for section in soup.findAll('a',href=True,attrs={'class':'_31qSD5'}):
        product_name=section.find('div',attrs={'class':'_3wU53n'})
        product_names.append(product_name.text)
        product_img=section.find('div',attrs={'class':'_3BTv9X'}).find('img')['src']
        if product_img[:4]=='http':
            product_imgs.append(product_img)
        product_price=section.find('div',attrs={'class':'_1vC4OE _2rQ-NK'})
        product_prices.append(product_price.text)
        product_rating=section.find('div',attrs={'class':'hGSR34'})
        product_ratings.append(product_rating.text)    
    #series=[elem.get_text() for elem in list1]
    #print(series)'''
```

```python
print("\n\nProduct Ratings:\n",product_ratings)
folder_path=input("\n\nEnter the directory path to store the images:\n\n")
for img_links in product_imgs:
    print(img_links)
    download_image(folder_path,img_links)
print("\n\nProduct Prices:\n",product_prices)
print("\n\nProduct Names:\n",product_names)
```

```python
data_frame=pd.DataFrame({'Product Name':product_names,'Price':product_prices,'Ratings':product_ratings})
data_frame.to_csv('Flipkart_products.csv',index=False,encoding='utf-8')
data_frame
```
